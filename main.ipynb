```python
import yaml
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
SAMPLE_TIME = 1000000
```

```python
def load_file(path) : 
    data = pd.read_csv(path)
    return data
```

## sigwait method

```python
execution_time = load_file("data/execution_time1.csv").to_numpy().squeeze()
execution_time_stress = load_file("data/execution_time2.csv").to_numpy().squeeze()

sample_time = load_file("data/sample_time1.csv").to_numpy().squeeze()
sample_time_stress = load_file("data/sample_time2.csv").to_numpy().squeeze()

idle_time = load_file("data/idle_time1.csv").to_numpy().squeeze()
idle_time_stress = load_file("data/idle_time2.csv").to_numpy().squeeze()
```

```python
sample_number = len(sample_time)
sample_numbe2 = len(sample_time_stress)
sample_number = range(sample_number)
sample_numbe2 = range(sample_numbe2)
plt.plot(sample_number, sample_time, color = "blue", label= "without stress")
plt.plot(sample_numbe2, sample_time_stress, color = "red", label= "with stress")
plt.legend()
plt.ylabel("period in nanoseconds")
plt.savefig("images/executiontimesig.png")
```

```python
plt.plot(range(len(execution_time)),execution_time, color='blue', label=   "without stress")
plt.plot(range(len(execution_time_stress)), execution_time_stress, label = "with stress", color = 'red')
plt.legend()
plt.ylabel("execution time in nanosecond")
plt.savefig("images/executiontimesig.png")
```
