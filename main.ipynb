```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
SAMPLE_TIME = 1000000
```

```python
def load_file(path) : 
    data = pd.read_csv(path)
    return data
```

## sigwait method

```python
execution_time = load_file("data/execution_time1.csv").to_numpy().squeeze()
execution_time_stress = load_file("data/execution_time2.csv").to_numpy().squeeze()

sample_time = load_file("data/sample_time1.csv").to_numpy().squeeze()
sample_time_stress = load_file("data/sample_time2.csv").to_numpy().squeeze()

idle_time = load_file("data/idle_time1.csv").to_numpy().squeeze()
idle_time_stress = load_file("data/idle_time2.csv").to_numpy().squeeze()
```

```python
"""statstics of the data """
mean_sample_time = np.mean(sample_time)
mean_sample_time_stress = np.mean(sample_time_stress)

std_sample = np.std(sample_time)
std_sample_stress = np.std(sample_time_stress)

mean_execution_time = np.mean(execution_time)
mean_execution_time_stress = np.mean(execution_time_stress)

std_execution_time = np.std(execution_time)
std_execution_time_stress = np.std(execution_time_stress)

print(f"mean sample time with out {mean_sample_time}")
print(f"mean sample time with stress {mean_sample_time_stress}")
print(f"std of sample time without {std_sample}")
print(f"std of sample with stress {std_sample_stress}")
print(f"mean execution time with out {mean_execution_time}")
print(f"mean execution time with stress {mean_execution_time_stress}")
print(f"std of execution time without {std_execution_time}")
print(f"std of execution with stress {std_execution_time_stress}")
```

```python
sample_number = len(sample_time)
sample_numbe2 = len(sample_time_stress)
sample_number = range(sample_number)
sample_numbe2 = range(sample_numbe2)
plt.plot(sample_number, sample_time, color = "blue", label= "without stress")
plt.plot(sample_numbe2, sample_time_stress, color = "red", label= "with stress")
plt.legend()
plt.ylabel("period in nanoseconds")
plt.savefig("images/sigwaitSample.png")
```

```python
plt.plot(range(len(execution_time)),execution_time, color='blue', label=   "without stress")
plt.plot(range(len(execution_time_stress)), execution_time_stress, label = "with stress", color = 'red')
plt.legend()
plt.ylabel("execution time in nanosecond")
plt.ylim(0,1.5e5)
plt.savefig("images/executiontimesig.png")
```

## xenomai

```python
execution_time = load_file("data3/execution_time1.csv").to_numpy().squeeze()
execution_time_stress = load_file("data3/execution_time2.csv").to_numpy().squeeze()

sample_time = load_file("data3/sample_time1.csv").to_numpy().squeeze()
sample_time_stress = load_file("data3/sample_time2.csv").to_numpy().squeeze()

idle_time = load_file("data3/idle_time1.csv").to_numpy().squeeze()
idle_time_stress = load_file("data3/idle_time2.csv").to_numpy().squeeze()
```

```python
"""statstics of the data """
mean_sample_time = np.mean(sample_time)
mean_sample_time_stress = np.mean(sample_time_stress)

std_sample = np.std(sample_time)
std_sample_stress = np.std(sample_time_stress)

mean_execution_time = np.mean(execution_time)
mean_execution_time_stress = np.mean(execution_time_stress)

std_execution_time = np.std(execution_time[1:100])
std_execution_time_stress = np.std(execution_time_stress[1:100])

print(f"mean sample time with out {mean_sample_time}")
print(f"mean sample time with stress {mean_sample_time_stress}")
print(f"std of sample time without {std_sample}")
print(f"std of sample with stress {std_sample_stress}")
print(f"mean execution time with out {mean_execution_time}")
print(f"mean execution time with stress {mean_execution_time_stress}")
print(f"std of execution time without {std_execution_time}")
print(f"std of execution with stress {std_execution_time_stress}")
```

```python
sample_number = len(sample_time)
sample_numbe2 = len(sample_time_stress)
sample_number = range(sample_number)
sample_numbe2 = range(sample_numbe2)
plt.plot(sample_number[1:100], sample_time[1:100],       color = "blue", label= "without stress")
plt.plot(sample_numbe2[1:100], sample_time_stress[1:100], color = "red", label= "with stress")
plt.ylabel("period in nanoseconds")
plt.legend()
plt.ylim(0.95e6,1.05e6)
plt.savefig("images/xenomaiSample.png")
```
